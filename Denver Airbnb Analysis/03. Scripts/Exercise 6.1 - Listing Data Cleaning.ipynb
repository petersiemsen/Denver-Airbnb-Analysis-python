## 2. Data Consistency Checks

### Column data types

```python
# Getting a better understanding of the columns and their data types 
df_listings.info()
```

```python
# Changing the columns with dates to the correct datetime format
df_listings['host_since'] = pd.to_datetime(df_listings['host_since'])
df_listings['first_review'] = pd.to_datetime(df_listings['first_review'])
df_listings['last_review'] = pd.to_datetime(df_listings['last_review'])
```

```python
# Converting 'price' datatype so it shows up in the summary statistics

# Remove dollar sign, commas, and extra spaces, then convert the column to numeric (float)
df_listings['price'] = df_listings['price'].str.replace('[\$, ]', '', regex=True).astype(float)
```

```python
# Updating all the values in the 'neighbourhood_cleansed' column for 'Gateway - Green Valley Ranch'
df_listings.loc[df_listings['neighbourhood_cleansed'] == 'Gateway - Green Valley Ranch', 'neighbourhood_cleansed'] = 'Gateway - GVR'
```

### Mixed-type data

```python
# Checking for mixed-type data
for col in df_listings.columns.tolist():
    weird = (df_listings[[col]].applymap(type) != df_listings[[col]].iloc[0].apply(type)).any(axis = 1)
    if len (df_listings[weird]) > 0:
        print (col)
```

```python
# Checking for mixed-type data and printing out the data type
for col in df_listings.columns.tolist():
    weird = (df_listings[[col]].applymap(type) != df_listings[[col]].iloc[0].apply(type)).any(axis=1)
    if weird.any():
        mixed_data = df_listings[weird]
        unique_data_types = mixed_data[col].apply(type).unique()
        print(f"Column '{col}' has mixed data types:")
        for dtype in unique_data_types:
            print(f"   - {dtype}")
```

```python
# Converting the following columns into string data types
df_listings[['host_location','host_response_time','host_response_rate','host_acceptance_rate','host_is_superhost',
             'host_neighbourhood','neighbourhood','bathrooms_text']] = df_listings[['host_location','host_response_time','host_response_rate','host_acceptance_rate','host_is_superhost',
             'host_neighbourhood','neighbourhood','bathrooms_text']].astype('str')
```

### Missing values

```python
# Checking for missing values
df_listings.isnull().sum()
```

#### I saw that the 'first_review' and 'last_review' columns have both mixed-type data as well as missing values. I am going to convert all the NaN values in these columns to an arbitrary start date of 01/01/1900. We can filter this out later in our analysis, but we need to make sure that we do not have mixed-type data before we begin.

```python
# Replacing the missing values in the two columns with '01/01/1900'
df_listings['first_review'] = df_listings['first_review'].fillna(pd.to_datetime('1900-01-01'))
df_listings['last_review'] = df_listings['last_review'].fillna(pd.to_datetime('1900-01-01'))
```

```python
# Checking to see if the missing values still exist
df_listings[['first_review','last_review']].isnull().sum()
```

```python
# Looking to see the values in the bathrooms_text column.
df_listings['bathrooms_text'].value_counts(dropna=False)
```
