```python
# Importing the 'listings_clean.pkl' and the 'calendar_clean.pkl' dataset
df_listings = pd.read_pickle(os.path.join(path, '02. Data\Modified\listings_clean.pkl'))

df_calendar = pd.read_pickle(os.path.join(path, '02. Data\Modified\calendar_clean.pkl'))
```

```python
df_listings
```

```python
# Changing the name of the id column so we can run the merge
df_listings.rename(columns = {'id' : 'listing_id'}, inplace = True)
```

```python
# Changing the name of columns in 'df_calendar' so there isn't any overlap
df_calendar.rename(columns = {'price' : 'daily_price'}, inplace = True)
df_calendar.rename(columns = {'adjusted_price' : 'daily_adjusted_price'}, inplace = True)
df_calendar.rename(columns = {'minimum_nights' : 'daily_min_nights'}, inplace = True)
df_calendar.rename(columns = {'maximum_nights' : 'daily_max_nights'}, inplace = True)
```

```python
df_calendar
```

## 2. Merging the datasets

```python
# Test merge without overwriting
pd.merge(df_calendar,df_listings, on = ['listing_id'], indicator = True)
```

#### The merge looks good. We are going to be dropping a lot of rows by using the inner join, but that is what we want. We will only want rows from the calendar dataset that have a listing_id and information associated with them (otherwise the analysis on those calendar rows would be useless). 

```python
# Running the actual merge, joining the listings dataframe on the calendar dataframe using the listing_id
df_cal_list =  df_calendar.merge(df_listings, on = 'listing_id', indicator = True)
```
