## 3. Additional charts to help answer questions from the project overview

### What is the distribution of Airbnb listing prices in Denver? 

```python
# Creating a histogram of the price column. Grouping the prices into bins first
bins = [0, 100, 200, 300, 400, 500, 600, 700, 800, 900, 1000, 20000, 68000]
bin_labels = ['0-100', '100-200', '200-300', '300-400', '400-500', '500-600', '600-700', '700-800', '800-900', '900-1000', '1000-20000', '20000-68000']

# Assigning the price values to a bin
df_listings['price_bins'] = pd.cut(df_listings['price'], bins=bins, labels=bin_labels, include_lowest=True)

# Setting the colors 
colors = sns.color_palette("Set2", 1)

# Set a grey background (use sns.set_theme() 
sns.set(style="darkgrid")

# Get the counts for each bin
bin_counts = df_listings['price_bins'].value_counts().sort_index()

# Plot the histogram using the counts and bin labels with align='edge'
plt.bar(range(len(bin_counts)), bin_counts, color=colors, align='edge', width=0.7)

# Set the tick positions and labels
plt.xticks(range(len(bin_counts)), bin_counts.index, rotation=45)

plt.xlabel('Price Groups')
plt.ylabel('Count')
plt.title('Histogram of Price Ranges')
plt.show()
```

#### The highest count of listing prices fall inbetween 100-200. The next highest count is from 0-100 dollars. The data is skewed left and the count tails off to the right: there are fewer high-priced listings and the price gets higher and higher. 

### Does the type of property significantly influence pricing?

```python
# Pivot the data to get property types as columns and price bins as index
grouped_data = df_listings.pivot_table(index='price_bins', columns='room_type', aggfunc='size', fill_value=0, sort=True)

# Plot the histogram for each property type
fig, ax = plt.subplots(figsize=(10, 6))

# Setting the colors 
colors = sns.color_palette("Set2", 7)

# Set a grey background (use sns.set_theme() 
sns.set(style="darkgrid")

grouped_data.plot(kind='bar', ax=ax, width = 1, color=colors, align='edge')

# Customize the plot
ax.set_ylabel('Count')
ax.set_title('Property Type Price Distribution')
ax.legend(title='Property Type')

# Set the tick positions and labels
plt.xticks(range(len(bin_counts)), bin_counts.index, rotation=45)

# Add count values on top of each bar (only if count is above 20)
for p in ax.patches:
    x = p.get_x() + p.get_width() / 2
    y = p.get_height()
    if y > 20:
        ax.annotate(str(int(y)), (x, y), ha='center', va='bottom')

# Show the plot
plt.tight_layout()
plt.show()
```
